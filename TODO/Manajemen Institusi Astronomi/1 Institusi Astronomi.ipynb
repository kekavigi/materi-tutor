# Institusi Astronomi Modern

Secara umum institusi astronomi formal dapat
dibagi menjadi 3 golongan besar
* Pendidikan Astronomi, contoh: department of astronomy
* Penelitian Astronomi, contoh: observatorium
* Astronomical Public Outreach, contoh: planetarium

## Jenis-jenis Institusi Astronomi

Selain institusi astronomi formal tersebut ada juga jenis yang kurang formal dan lebih cair, tapi dampaknya terhadap perkembangan astronomi tidak kalah dengan institusi formal yaitu

* Perkumpulan astronomi, contoh : klub astronom amatir / profesional
* Bisnis astronomi, contoh : badan usaha bertema astronomi
* Institusi virtual contoh : virtual observatory, astronomical network

Sebenarnya pembagian diatas tidak benar-benar terpisah, umumnya institusi astronomi melakukan lebih dari satu peran dari enam jenis institusi diatas, namun penekanannya berbeda. Kadang pada suatu institusi, kegiatan pendidikan dan penelitiannya cukup seimbang sehingga dapat dikategorikan lembaga pendidikan maupun penelitian.

* Ada lembaga penelitian yang mempunyai observatorium virtual
* Ada bisnis yang membina klub astronomi
* Ada klub yang mempunyai network yang luas.
* Ada lembaga pendidikan yang melakukan public outreach
* Dan lain-lain.

## Institusi Pendidikan

Institusi astronomi jenis ini adalah yang kegiatannya terutama mendidik calon-calon astronom, tapi juga melakukan pengamatan dan penelitian, kadang-kadang juga melayani masyarakat. Contohnya adalah jurusan-jurusan astronomi di universitas seperti prodi astronomi ITB, Department of Astronomy Kyoto University dll. Bisa juga institusi pendidikan astronomi merupakan subjurusan di jurusan lain seperti Fisika, Geosains, Matematika dll. Misalnya, di Yukawa Institute of Theoretical Physics, Department of Physics, Kyoto University, ada divisi astrofisika. Ciri institusi pendidikan adalah mempunyai
kurikulum yang dipakai untuk mendidik calon astronom, dan memberikan diploma kepada lulusannya.

## Institusi Penelitian

Institusi jenis ini kegiatan utamanya adalah penelitian, akan tetapi umumnya juga melakukan kegiatan pendidikan dan juga layanan untuk masyarakat. Institusi penelitian murni kegiatan utamanya adalah meneliti, tidak mempunyai kurikulum dan tidak memberikan diploma. Contoh institusi kategori ini adalah observatorium (misalnya Observatorium Bosscha, Observatorium Gunma di Jepang) dan Institut Astronomi, (misalnya Kapteyn Astronomical Institute,

## Perkumpulan Astronomi

Dapat dibagi menjadi 2 golongan besar:
* Profesional, perkumpulan orang-orang yang menjadikan astronomi menjadi bidang profesi. Biasanya anggotanya mendapat pendidikan formal asrronomi dan bekerja di institusi formal astronomi, contoh HAI, IAU
* Amatir, perkumpulan orang-orang yang hobby astronomi. Biasanya pendidikan formal anggotanya bukan astronomi. Contoh: HAAJ, klub astronomi sekolah, VSOLJ

## Lembaga Virtual

Berperan di dunia maya (cyber), memberikan kontribusi penting kepada dunia astronomi. Contoh : VSNET

Virtual Observatory, memberikan layanan data melalui internet dengan berbagai informasi yang biasa diperoleh di observatorium, seperti citra, tabel numerik, spektrum, paper dll. Contoh : Sloan Digital Sky Survey (SDSS).

Jaringan (network) astronomi : hubungan antar manusia bertema astronomi. Di dalam network ini, anggota saling bertukar informasi, bertukar data, berdiskusi. Biasanya jaringan cyber ini berafiliasi pada klub, tapi ada juga yang tidak. Contoh: VSNET (Variable Star Network). Meskipun network ini tidak mempunyai basis secara fisik, tapi kontribusinya bisa melebihi institusi formal.

## Proses Manajemen Pada Institusi Astronomi

* **Planing**: Renstra, Rencana Tahunan, Budget, rencana cadangan
* **Organizing**: Organigram, SOP, Financial flow, information flow, administrative mechanism
* **Leading**: Menjalankan operasional institusi untuk mencapai target dan tujuan, scientific/academic leadership
* **Controlling**: Pengawasan, seminar, proposal review, audit, tindakan korektif, board of director meeting.
