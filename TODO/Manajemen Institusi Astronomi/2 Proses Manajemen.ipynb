# Proses Manajemen

## Organisasi

Pengertian: Kumpulan manusia yang bekerja bersama-sama
untuk mencapai tujuan bersama.

Organisasi menghasilkan produk atau jasa yang bernilai bagi masyarakat dan memenuhi kebutuhan konsumen/stakeholders

* Organisasi merupakan sistem terbuka yang berisi bagian-bagian yang berfungsi bersama-sama untuk mencapai tujuan bersama
* Berinteraksi dengan lingkungan
* Mentransformasikan sumber daya input menjadi produk atau jasa
* Lingkungan memberi umpan balik yang mencerminkan sebaik apa organisasi itu memenuhi kebutuhan konsumen dan masyarakat
* Umpan balik itu membuat organisasi hidup

## Kinerja Organisasi

"Penciptaan nilai: merupakan wacana yang sangat penting bagi organisasi Nilai diciptakan ketika organisasi mengubah atau menambah nilai pada biaya yang diinputkan Ketika terjadi penciptaan nilai :
* Bisnis mendapat laba
* Organisasi nirlaba menambah kekayaan pada masyarakat

* Produktivitas : ukuran umum kinerja secara kualitatif maupun kuantitatif, dengan memperhitungkan penggunaan sumber daya
* Efektivitas : ukuran pencapaian pada output
* Efisiensi : Ukuran pemakaian sumber daya pada input untuk mencapai target

## Pentingnya sumber daya manusia
Orang bukanlah biaya yang harus ditekan serendah mungkin. Organisasi yang berkinerja tinggi memperlakukan
manusia sebagai aset berharga yang strategis. Manager harus memastikan bahwa orang diperlakukan sebagai aset strategis

## Manager
Mendukung langsung dan membantu mengaktifkan upaya-upaya pihak lain untuk mencapai kinerja terbaik. Orang-orang yang dibantu manager adalah mereka. yang tugasnya merupakan pekerjaan nyata organisasi

## Tingkatan Manajemen

* Top manager bertanggung jawab terhadap kinerja organisasi secara umum.
* Middle manager bertanggung jawab terhadap kinerja suatu bagian atau divisi yang cukup besar.
* Team leader atau supervisor bertanggung jawab terhadap kinerja suatu kelompok kecil pekerja non managerial

## Jenis manager

* Line manager mengelola langsung pekerja yang menghasilkan produk organisasi
* Staff manager menggunakan kepakaran teknis untuk membantu pekerja
* Functional manager bertanggung jawab terhadap suatu jenis aktivitas organisasi
* General manager bertanggung jawab terhadap unit yang lebih kompleks yang melibatkan berbagai fungsi
* Administrator adalah manager di organaisasi non profit atau publik

## Quality of work life (QWL)

Merupakan indikator kualitas manusia di tempat kerja. Indikator itu meliputi:
* Gaji yang memadai
* Kondisi kerja yang aman
* Kesempatan untuk belajar dan menggunakan ketrampilan baru
* Kemungkinan untuk mengembangkan karir
* Perlindungan hak individual
* Kebanggaan terhadap pekerjaan yang dilakukan

## Manager yang berhasil

* Mengetahui kebutuhan para supervisor/ team leader.
* Memberikan dukungan dan saran kepada bawahannya.
* Membantu bawahannya berprestasi optimum

## Akuntabilitas

Akuntabilitas adalah keharusan untuk bertanggung jawab kepada otoritas yang lebih tinggi tentang kinerja yang menjadi tugasnya Manager yang efektif memenuhi akuntabilitas kinerja dengan cara membantu orang lain mencapai kinerja tinggi dan mencapai kepuasan dalam bekerja

## Management process

Manager membuat organisasinya berkinerja tinggi dengan menggunakan orang2 dan material di bawahnya secara optimal. Manajemen adalah proses perencanaan, pengorganisasian, kepemimpinan dan pengendalian sumberdaya untuk mencapai tujuan. Manager bertanggung jawab atas keempat fungsi itu. Fungsi itu dilaksanakan secara kontinu

## Peran managerial

* Peran interpersonal : interaksi dengan orang2 di dalam dan di luar organisasi
* Peran informasi : mencari, memberikan dan menganalisis informasi
* Peran pengambilan keputusan : menggunakan informasi yang didapat dan dianalisis untuk membuat keputusan untuk menyelesaikan masalah atau meraih kesempatan

## Ketrampilan pokok managerial

* Ketrampilan : kemampuan untuk menerjemahkan pengetahuan menjadi tindakan yang menghasilkan kinerja yang diinginkan
* Ketrampilan teknis : kemampuan menerapkan suatu keahlian untuk melakukan suatu pekerjaan khusus
* Ketrampilan interpersonal : kemampuan untuk bekerja sama dengan orang lain.
* Ketrampilan konseptual : kemampuan berfikir kritis dan analitis untuk mengatasi masalah2 komplex

## Agenda setting dan networking

* Agenda setting : menetapkan prioritas tindakan. Agenda setting juga termasuk menetapkan target dan rencana jangka pendek dan jangka panjang
* Networking adalah proses pengembangan dan pemeliharaan hubungan positif dengan pihak-pihak yang dapat membantu pelaksanaan tugas

## Kompetensi managerial

* Komunikasi
* Team work
* Inisiatif
* Manajemen diri
* Kepemimpinan
* Kemampuan berfikir kritis
* Profesionalisme

Kompetensi managerial secara implisit sudah termasuk kemampuan2:
* Planning, organizing, leading dan controlling
* Interpersonal, informational, pengambilan keputusan
* Agenda setting dan networking
